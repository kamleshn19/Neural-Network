Problem Statement : To Predict the churning rate for employee in company

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```
#Load Dataset and split into train and test
employeeChurn = pd.read_csv("/content/drive/My Drive/Colab Notebooks/DeepLEarning/Churn_Data raw data-1.csv")
featureVariables = employeeChurn.iloc[:,4:19]
targetVariables = employeeChurn.iloc[:,20]
```

```
#Create dummy variables for categorical features
IntPlan = pd.get_dummies(featureVariables[" Int'l Plan"],drop_first=True)
VmailPlan = pd.get_dummies(featureVariables[" VMail Plan"],drop_first=True)
```

```
#Concatinate new dummy varibles and remove original categorical feature variables
featureVariables = pd.concat([featureVariables,IntPlan,VmailPlan],axis=1)
featureVariables = featureVariables.drop([" Int'l Plan"," VMail Plan"],axis=1)
```

```
#Splitting dataset in training and test
xTrain , xTest, yTrain, yTest = train_test_split(featureVariables,targetVariables, test_size=0.3, random_state=522)
```

```
# Feature Scaling
sc = StandardScaler()
xTrain = sc.fit_transform(xTrain)
xTest = sc.fit_transform(xTest)
```

```
# Initialise the ANN
classifier = Sequential()
#Add Hidden Layer
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=15))
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```
#complaining ANN classifier
classifier.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])
```

```
model = classifier.fit(xTrain,yTrain,batch_size=5, nb_epoch=100)
```

```
#Predicting on test dataset
yPredicted = classifier.predict(xTest)
yPredicted = yPredicted>0.5
```

```
testScore =accuracy_score(yPredicted,yTest)
```

```
print("Accuracy Test Dataset:",testScore*100,"%")
```
